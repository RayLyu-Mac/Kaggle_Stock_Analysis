```python
import polars as pl
from os import listdir
import matplotlib.pyplot as plt
import numpy as np
```

```python
for file in listdir("CAData\\"):
    df=pl.read_parquet(f"CAData\\{file}")
    price=df["Price"].to_numpy()
    price=(price-np.mean(price))/np.std(price)
    plt.plot(price)
```
